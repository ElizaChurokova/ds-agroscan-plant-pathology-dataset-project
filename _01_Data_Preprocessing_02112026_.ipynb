```python
import pandas as pd
import os
import shutil
from sklearn.model_selection import train_test_split

photos_path = '/content/drive/MyDrive/my_photos'
csv_path = '/content/drive/MyDrive/train.csv'
split_dir = '/content/drive/MyDrive/plant_dataset_split'

df = pd.read_csv(csv_path)

# Определяем 5 четких целей для презентации
target_labels = ['healthy', 'scab', 'rust', 'frog_eye_leaf_spot', 'powdery_mildew']

for label in target_labels:
    for folder in ['train', 'val', 'test']:
        os.makedirs(os.path.join(split_dir, folder, label), exist_ok=True)

print("Nachinayu umnuyu sortirovku po 5 osnovnym klassam...")

for target in target_labels:
    # Ishem vse stroki, gde vstrechaetsya nuzhnoe slovo
    # Naprimer, 'scab frog_eye' popadet i v scab, i v frog_eye (ili v odno iz nih)
    matched_images = df[df['labels'].str.contains(target)]['image'].tolist()

    existing_images = [img for img in matched_images if os.path.exists(os.path.join(photos_path, img))]

    if len(existing_images) > 0:
        train_imgs, temp_imgs = train_test_split(existing_images, test_size=0.2, random_state=42)
        val_imgs, test_imgs = train_test_split(temp_imgs, test_size=0.5, random_state=42)

        for imgs, folder in zip([train_imgs, val_imgs, test_imgs], ['train', 'val', 'test']):
            for img in imgs:
                shutil.copy(os.path.join(photos_path, img), os.path.join(split_dir, folder, target, img))

        print(f"Class {target}: gotovo, razmesheno {len(existing_images)} foto.")
    else:
        print(f"Vnimanie: Dlya klassa {target} faily ne naideny.")

print("\nSortirovka zavershena. Teper u nas est vse 5 klassov!")
```

```python
import os
import shutil

dataset_path = '/content/drive/MyDrive/plant_dataset_split'
target_labels = ['healthy', 'scab', 'rust', 'frog_eye_leaf_spot', 'powdery_mildew']

for folder in ['train', 'val', 'test']:
    folder_path = os.path.join(dataset_path, folder)
    # Spisok vseh papok vnutri
    current_dirs = os.listdir(folder_path)

    for d in current_dirs:
        if d not in target_labels:
            full_path = os.path.join(folder_path, d)
            shutil.rmtree(full_path)
            print(f"Udalena lishnyaya papka: {d}")

print("Teper v datashite tolko 5 nuzhnyh klassov.")
```
